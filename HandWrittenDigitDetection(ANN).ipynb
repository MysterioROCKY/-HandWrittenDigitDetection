```python
new_X_train = X_train/255
new_X_test = X_test/255
```

```python
new_X_train[0][5]
```

```python
new_flat_X_train = new_X_train.reshape(len(new_X_train),28*28)
new_flat_X_test = new_X_test.reshape(len(new_X_test),28*28)
```

```python
model = keras.Sequential([
                          keras.layers.Dense(units=10,
                                             input_shape=(784,),
                                             activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(new_flat_X_train,y_train, epochs=5)
```

```python
model.evaluate(new_flat_X_test,y_test)
```

# # Therefore, the Accuracy of our model is 92%

```python
y_pred = model.predict(new_flat_X_test)
```

```python
y_pred
```

```python
y_pred[0]
```

```python
# argmax will give the index of maximum value which is essential the digit, because this array includes the probability of 
# digits from 0 to 9.
np.argmax(y_pred[0])
```

```python
y_pred_label = [np.argmax(i) for i in y_pred]
```

```python
y_pred_label
```

```python
# Confusion Matrix , shows how many times model predict correct value
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_label)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
```

# Multiple Hidden Layes
